```python
def predict(Theta1, Theta2, X):
    m = X.shape[0]
    num_labels = Theta2.shape[0]
    p = np.zeros(m)
    h1 = sigmoid(np.dot(np.concatenate([np.ones((m, 1)), X], axis=1), Theta1.T))
    h2 = sigmoid(np.dot(np.concatenate([np.ones((m, 1)), h1], axis=1), Theta2.T))
    p = np.argmax(h2, axis=1)
    for i in range(h2.size):
        if(h2[i]<0.5):
            p[i]=0
        else:
            p[i]=1
    return p


def debugInitializeWeights(fan_out, fan_in):
    # Initialize W using "sin". This ensures that W is always of the same values and will be
    # useful for debugging
    W = np.sin(np.arange(1, 1 + (1+fan_in)*fan_out))/10.0
    W = W.reshape(fan_out, 1+fan_in, order='F')
    return W


def computeNumericalGradient(J, theta, e=1e-4):
    numgrad = np.zeros(theta.shape)
    perturb = np.diag(e * np.ones(theta.shape))
    for i in range(theta.size):
        loss1, _ = J(theta - perturb[:, i])
        loss2, _ = J(theta + perturb[:, i])
        numgrad[i] = (loss2 - loss1)/(2*e)
    return numgrad


def checkNNGradients(nnCostFunction, lambda_=0):
    input_layer_size = 3
    hidden_layer_size = 5
    num_labels = 3
    m = 5

    # We generate some 'random' test data
    Theta1 = debugInitializeWeights(hidden_layer_size, input_layer_size)
    Theta2 = debugInitializeWeights(num_labels, hidden_layer_size)

    # Reusing debugInitializeWeights to generate X
    X = debugInitializeWeights(m, input_layer_size - 1)
    y = np.arange(1, 1+m) % num_labels
    # Unroll parameters
    nn_params = np.concatenate([Theta1.ravel(), Theta2.ravel()])

    # short hand for cost function
    costFunc = lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size,
                                        num_labels, X, y, lambda_)
    cost, grad = costFunc(nn_params)
    numgrad = computeNumericalGradient(costFunc, nn_params)

    # Visually examine the two gradient computations.The two columns you get should be very similar.
    print(np.stack([numgrad, grad], axis=1))
    print('The above two columns you get should be very similar.')
    print('(Left-Your Numerical Gradient, Right-Analytical Gradient)\n')

    # Evaluate the norm of the difference between two the solutions. If you have a correct
    # implementation, and assuming you used e = 0.0001 in computeNumericalGradient, then diff
    # should be less than 1e-9.
    diff = np.linalg.norm(numgrad - grad)/np.linalg.norm(numgrad + grad)

    print('If your backpropagation implementation is correct, then \n'
          'the relative difference will be small (less than 1e-9). \n'
          'Relative Difference: %g' % diff)


def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
input_layer_size  = 30 # 30 input features
hidden_layer_size = 50   # 50 hidden units
num_labels = 1          # 1 label, 0: B, 1: M
```

```python
def nnCostFunction(nn_params,
                   input_layer_size,
                   hidden_layer_size,
                   num_labels,
                   X, y, lambda_=0.0):
  
    # Reshape nn_params back into the parameters Theta1 and Theta2, the weight matrices
    # for our 2 layer neural network
    Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                        (hidden_layer_size, (input_layer_size + 1)))

    Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                        (num_labels, (hidden_layer_size + 1)))

    m = y.size
         
    J = 0
    Theta1_grad = np.zeros(Theta1.shape)
    Theta2_grad = np.zeros(Theta2.shape)

    h = np.zeros(m,)
    X = np.concatenate([np.ones((m, 1)), X], axis=1)
    del1=0
    del2=0
    k=num_labels
    #ny=np.zeros([m,k])
    #nh=np.zeros([m,k])
    for i in range(m):
        a1 = X[i][:,None] 
        z2=np.dot(Theta1,a1)
        a2=sigmoid(z2)
        a2 = np.vstack((np.ones(1),a2))
        z3=np.dot(Theta2,a2)
        h=sigmoid(z3)
        t=int(y[i])
        #ny[i,t]=1
        c=(np.sum(-(np.multiply(int(y[i]),np.log(h)))-(np.multiply((1-int(y[i])),np.log(1-h)))))/m
        J=J+c
        
        d3=h-y[i]
        d2=np.dot(np.transpose(Theta2),d3) 
        d2=d2[1:] * sigmoidGradient(z2)
        del2=del2+ np.dot(d3,np.transpose(a2))
        del1=del1+ np.dot(d2,np.transpose(a1))
        
    t1=np.sum(np.square(Theta1[:,1:]))
    t2=np.sum(np.square(Theta2[:,1:]))
    reg=lambda_ * (t1+t2) / (2*m) 
    J=J+reg
    Theta1_grad[:,1:] = del1[:,1:]/m + lambda_ * (Theta1[:,1:]) /m
    Theta2_grad[:,1:] = del2[:,1:]/m + lambda_ * (Theta2[:,1:]) /m 
    Theta1_grad[:,0] = del1[:,0]/m
    Theta2_grad[:,0] = del2[:,0]/m   
    
    grad = np.concatenate([Theta1_grad.ravel(), Theta2_grad.ravel()])

    return J, grad
```

```python
def sigmoidGradient(z):

    g = np.zeros(z.shape)
    s=sigmoid(z)
    g= s * ( 1-s )

    return g
```

```python
def randInitializeWeights(L_in, L_out, epsilon_init=0.12):
    W = np.zeros((L_out, 1 + L_in))
    W = ( np.random.rand(L_out, 1 + L_in) * 2 * epsilon_init )- epsilon_init
    return W
```

```python
print('Initializing Neural Network Parameters ...')

initial_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)
initial_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)

# Unroll parameters
initial_nn_params = np.concatenate([initial_Theta1.ravel(), initial_Theta2.ravel()], axis=0)
```

```python
checkNNGradients(nnCostFunction)
```

```python
#  Check gradients by running checkNNGradients
lambda_ = 3
checkNNGradients(nnCostFunction, lambda_)
```

```python
options= {'maxiter': 200}

lambda_ = 0.5

# Create "short hand" for the cost function to be minimized
costFunction = lambda p: nnCostFunction(p, input_layer_size,
                                        hidden_layer_size,
                                        num_labels, X, y, lambda_)

# Now, costFunction is a function that takes in only one argument
# (the neural network parameters)
res = optimize.minimize(costFunction,
                        initial_nn_params,
                        jac=True,
                        method='TNC',
                        options=options)

# get the solution of the optimization
nn_params = res.x
        
# Obtain Theta1 and Theta2 back from nn_params
Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                    (hidden_layer_size, (input_layer_size + 1)))

Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                    (num_labels, (hidden_layer_size + 1)))
```

```python
pred = predict(Theta1, Theta2, X)
print('Training Set Accuracy: %f' % (np.mean(pred == y) * 100))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.4, random_state=42)
X_cv, X_test, y_cv, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=42)
```

```python
l=[0,0.0001,0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]
l2=[]

for i in range(12):
    mlp=  MLPClassifier(hidden_layer_sizes=(50,), max_iter=200, alpha=l[i],
                    solver='sgd', verbose=0, tol=1e-4, random_state=1,
                    learning_rate_init=.1)

    mlp.fit(X_train, y_train)
    l2.append(mlp)
    print(l[i],"\t\t",  mlp.score(X_train, y_train), "\t\t" , mlp.score(X_cv, y_cv))
```

```python
# accuracy on test set for alpha with highest accuracy on cross validation set 
for i in range(10,12):
    print(l[i],"\t",  l2[i].score(X_test, y_test))
```
