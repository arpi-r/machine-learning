#### Import required libraries

```python
import numpy as np
import pandas as pd
```

#### Load dataset

```python
data=pd.read_csv("data.csv")
data.head()
```

#### Label Encode the output

This makes the diagnosis M: 1 and B:0

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data.diagnosis=le.fit_transform(data.diagnosis)
data.head()
```

#### Remove unnecessary features

```python
data = data.drop(["id","Unnamed: 32"],axis=1)
```

```python
y=data.diagnosis
y.tail()
```

```python
x=data.drop("diagnosis",axis=1)
x.head()
```

#### Standardization

```python
x = (x - x.mean()) / (x.std())   
```

#### Import machine learning models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

#### Stacked Model

Split the dataset

```python
X_train, X_validation, Y_train, Y_validation = train_test_split(x, y, test_size=0.30, random_state=0)
```

Random forest model

Creating a new dataset with predictions from the random forest model

```python
X = np.zeros((X_validation.shape[0],31))
X.shape
X[:,0:30]=X_validation
```

```python
rf=RandomForestClassifier(random_state=3)
rf.fit(X_train, Y_train)
predictions = rf.predict(X_validation)
X[:,30]=predictions
print(accuracy_score(Y_validation, predictions))
```

Split the new dataset

```python
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y_validation, test_size=0.30, random_state=0)
```

Logistic regression model

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
predictions = lr.predict(X_validation)
print(accuracy_score(Y_validation, predictions))
```

#### Stacked model using train, cross validation and test datasets

Split the dataset into 3, train, validation and test

```python
X_train, X_other, y_train, y_other = train_test_split(x, y, test_size=0.4, random_state=0)
X_cv, X_test, y_cv, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=0)
```

```python
X = np.zeros((X_cv.shape[0],31))
X.shape
X[:,0:30]=X_cv
```

Random forest model

creating random forest models with different random_state s

```python
l=[]
for i in range(100):
    print(i)
    rf=RandomForestClassifier(random_state=i)
    l.append(rf)
    rf.fit(X_train, y_train)
    predictions = rf.predict(X_cv)
    print(accuracy_score(y_train, rf.predict(X_train)))
    print(accuracy_score(y_cv, predictions))
```

Random state = 97 gives highest cross validation accuracy, so creating a new dataset with the predictions from this random forest model

```python

predictions = l[97].predict(X_cv)
print(accuracy_score(y_train, l[97].predict(X_train)))
print(accuracy_score(y_cv, predictions))
X[:,30]=predictions
print(accuracy_score(y_test, l[97].predict(X_test)))
```

Split the new dataset into 3

```python
X_train, X_other, y_train, y_other = train_test_split(X, y_cv, test_size=0.4, random_state=0)
X_cv, X_test, y_cv, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=0)
```

Logistic Regression model

Creating logistic regression models with different C

```python
l=[1/0.0001,1/0.0003, 1/0.001, 1/0.003, 1/0.01, 1/0.03, 1/0.1, 1/0.3, 1, 1/3, 1/10]
l2=[]
for i in range (11):
    print(l[i])
    lr = LogisticRegression(C=l[i])
    l2.append(lr)
    lr.fit(X_train, y_train)
    predictions = lr.predict(X_cv)
    print(accuracy_score(y_train, lr.predict(X_train)))
    print(accuracy_score(y_cv, predictions))
    print()
```

C= 1/0.003, 1/0.01, 1/0.03, 1/0.1, 1/0.3, 1   give highest validation accuracy, so testing these models with the test dataset

```python


for i in range (3,9):
    print(l[i])
    lr=l2[i]
    predictions = lr.predict(X_cv)
    print(accuracy_score(y_train, lr.predict(X_train)))
    print(accuracy_score(y_cv, predictions))
    print(accuracy_score(y_test, l2[9].predict(X_test)))
    print()
```

Accuracy on test dataset=1.
